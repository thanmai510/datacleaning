```python
#deleting empty cell
import pandas as pd
from pandas import read_csv
path="/content/drive/MyDrive/rawdata.csv"
df=pd.read_csv(path)
new_df=df.dropna()
print(new_df.to_string())
```

```python
#replacing the empty cells
new_df=df.fillna(130)
print(new_df.to_string())
```

```python
#replace only for specified coulunm
new_df=df["Duration"].fillna(60)
print(new_df.to_string())
```

```python
#replace only for specified coulunm
new_df=df["Calories"].fillna(130)
print(new_df.to_string())
```

```python
#replace using mean
x=df["Duration"].mean()
new_df=df["Duration"].fillna(x);
print(x)
print(new_df.to_string())
```

```python
#replace using mode
z=df["Maxpulse"].mode()
new_df=df["Maxpulse"].fillna(z);
print(z)
print(new_df.to_string())
```

```python
#replace using median
y=df["Pulse"].median()
new_df=df["Pulse"].fillna(y);
print(y)
print(new_df.to_string())
```

```python
#fixing wrong data
df.loc[7,'Duration']=45
print(df.to_string())
```

```python
#using loop replacing particular value in column
for x in df.index:
  if df.loc[x,"Duration"]>120:
    df.loc[x,"Duration"]=120
print(df.to_string())
```
